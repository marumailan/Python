Write a function that takes in a string of one or more words, and returns the same string, but with all words that have five or more letters reversed (just like the name of this kata). Strings passed in will consist of only letters and spaces. Spaces will be included only when more than one word is present.

Examples:

"Hey fellow warriors"  --> "Hey wollef sroirraw" 
"This is a test        --> "This is a test" 
"This is another test" --> "This is rehtona test"

```python
sentence = "Hey fellow warriors"

words = sentence.split()
result = []

for char in words:
    char_counts = (len(char))

    if (char_counts) >= 5:
        result.append(char[::-1])
    else:
        result.append(char)

print(" ".join(result))


```

```python
def spin_words(sentence):
     
    words = sentence.split()
    result = []

    for char in words:
        char_counts = (len(char))

        if (char_counts) >= 5:
            result.append(char[::-1])
        else:
            result.append(char)
    
    return " ".join(result)
```

We need a function that can transform a number (integer) into a string.

What ways of achieving this do you know?

Examples (input --> output):

123  --> "123"
999  --> "999"
-100 --> "-100"

```python
def number_to_string(num):
    return str(num)
```

Explanation

print vs return — the key difference:

print — displays a value to the screen. That's it. The value is gone after that.
return — sends a value back to the caller, so it can be stored, reused, or passed to another function.


Example showing the difference:
python# ❌ Using print — the result is lost after display
def add_with_print(a, b):
    print(a + b)  # shows it, but gives nothing back

result = add_with_print(3, 4)  # prints 7
print(result)                  # prints None ← nothing was returned!


# ✅ Using return — the result can be reused
def add_with_return(a, b):
    return a + b  # sends the result back to the caller

result = add_with_return(3, 4)  # stores 7 in `result`
print(result)                   # prints 7

Applying this to your number_to_string problem:
python# ❌ What you did — prints but returns nothing useful
def number_to_string(num):
    print(str(num))

output = number_to_string(123)
print(output)        # None ← can't use the value later


# ✅ What you should do — return the value
def number_to_string(num):
    return str(num)

output = number_to_string(123)
print(output)        # "123" ← now you can use it!

The mental model:
Think of a function like a vending machine:

print is like the machine showing you a picture of a snack on a screen — you can see it but can't eat it.
return is the machine actually dispensing the snack — you get something you can use.


One more thing — pass:
pass is a no-op placeholder that does nothing at all. It's used when Python requires a code block but you have nothing to put there yet:
pythondef number_to_string(num):
    pass  # ← function exists but does nothing, returns None
So the hierarchy is: return > print > pass in terms of usefulness for getting a value out of a function.

In this simple assignment you are given a number and have to make it negative. But maybe the number is already negative?

Examples

make_negative(1);  # return -1
make_negative(-5); # return -5
make_negative(0);  # return 0
Notes

The number can be negative already, in which case no change is required.
Zero (0) is not checked for any specific sign. Negative zeros make no mathematical sense.

```python
def make_negative( number ):
    if number < 0:
        return number
    else:
        return number * -1
```

Your classmates asked you to copy some paperwork for them. You know that there are 'n' classmates and the paperwork has 'm' pages.

Your task is to calculate how many blank pages do you need. If n < 0 or m < 0 return 0.

Example:

n= 5, m=5: 25
n=-5, m=5:  0

```python
def paperwork(n, m):
    if n > 0 and m > 0:
        return n * m
    else:
        return 0
```
