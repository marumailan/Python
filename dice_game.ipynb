```python
A simple dice game generated by random numbers by two die
```

```python
import random

dice_1 = random.randint(1,6)
dice_2 = random.randint(1,6)

total_dice = dice_1 + dice_2
 
print("die one is: " + str(dice_1) + "\nand\n" + "die two is: " + str(dice_2), "\nTotal output is: " + str(total_dice))
```
